```python
from rich.console import Console
from rich.table import Table
import sqlite3

conn = sqlite3.connect(str('data/default/stats_all.db'))
cur = conn.cursor()

# with open("eval.sql", "rt") as f:
#     cur.executescript(f.read())
```

```python
def print_table(title, query):
    res = cur.execute(query)
    rows = (res.fetchall())

    table = Table(title=title, width=300, box=None, expand=True)
    for header in res.description:
        table.add_column(header[0])
    for row in rows:
        table.add_row(*(str(rr) for rr in row))
    console = Console()
    console.print(table)

print("="*200)

print_table("Dev", """

SELECT *
FROM all_run_results
limit 100

""")
```
